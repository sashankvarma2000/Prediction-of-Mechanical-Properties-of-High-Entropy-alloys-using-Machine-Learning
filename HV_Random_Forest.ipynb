```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv ('/content/HV26.xlsx - Sheet1.csv')
```

```python
df
```

```python
df.drop(['Unnamed: 0'], axis = 1, inplace = True)
```

```python
df['BulkModulus'] = df['BulkModulus'].str.replace(',', '')
```

```python
df['Tm'] = df['Tm'].str.replace(',', '')


```

```python
df = pd.get_dummies(df, columns = ['Type of phases','Phase','Phase_inshort'])
```

```python
df=df.astype(float)
```

```python
df
```

```python
y=df["HV"]
```

```python
df.drop("HV",axis=1,inplace=True)
```

```python
# copy the data
df_max_scaled = df.copy()

# apply normalization techniques
for column in df_max_scaled.columns:
	df_max_scaled[column] = df_max_scaled[column] / df_max_scaled[column].abs().max()
	
# view normalized data
display(df_max_scaled)
```

```python
df=df_max_scaled
```

```python
df.drop("E# (GPa)",axis=1,inplace=True)
```
