```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv ('/content/HV26.xlsx - Sheet1.csv')
```

```python
df
```

```python
df.columns
```

```python
df.drop(['Unnamed: 0'], axis = 1, inplace = True)
```

```python
df['BulkModulus'] = df['BulkModulus'].str.replace(',', '')
```

```python
df['Tm'] = df['Tm'].str.replace(',', '')


```

```python
df = pd.get_dummies(df, columns = ['Type of phases','Phase','Phase_inshort'])
```

```python
df=df.astype(float)
```

```python
df
```

```python
y=df["E# (GPa)"]
```

```python
# copy the data
df_max_scaled = df.copy()

# apply normalization techniques
for column in df_max_scaled.columns:
	df_max_scaled[column] = df_max_scaled[column] / df_max_scaled[column].abs().max()
	
# view normalized data
display(df_max_scaled)
```

```python
df=df_max_scaled
```

```python
df.drop("HV",axis=1,inplace=True)
```

```python
X_train,y_train=df[0:71],y[0:71]
```

```python
x_test,y_test=df[71:96],y[71:96]
```

```python
X_train.drop(["E# (GPa)"],axis = 1, inplace = True)
```

```python
x_test.drop(["E# (GPa)"],axis=1,inplace=True)
```

```python
from sklearn.neural_network import MLPRegressor
```

```python
regr = MLPRegressor(random_state=1, max_iter=2000).fit(X_train, y_train)
```

```python
pred=regr.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error as mse
```
